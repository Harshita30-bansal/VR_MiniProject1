This notebook uploads the `VRMP1_<roll_no>` folder (containing `predictor.py`, `validator_local.py`, `Student_Submission_Guidelines.md`, and `model_files/`) to a Hugging Face repository.

## 2. Authenticate with Hugging Face

Generate a **write-access token** at https://huggingface.co/settings/tokens, then paste it below.

```python
from huggingface_hub import login

# Paste your Hugging Face write-access token here
HF_TOKEN = ""   # <-- replace with your actual token

login(token=HF_TOKEN)
```

## 3. Create a New Repository

Choose a `repo_type`: `"model"` (for `.pth` weights + code) or `"dataset"`.

```python
from huggingface_hub import HfApi

api = HfApi()

# ── Configure these ──────────────────────────────────────────────
HF_USERNAME  = "harshita3030"   # <-- your HF username
REPO_NAME    = "VRMP1_IMT2023035_IMT2023537_IMT2023550_IMT2023585"   # <-- desired repo name
REPO_TYPE    = "model"              # "model" | "dataset" | "space"
PRIVATE      = False                 # True = private repo
# ─────────────────────────────────────────────────────────────────

REPO_ID = f"{HF_USERNAME}/{REPO_NAME}"

api.create_repo(
    repo_id=REPO_ID,
    repo_type=REPO_TYPE,
    private=PRIVATE,
    exist_ok=True,   # won't fail if the repo already exists
)

print(f"Repository '{REPO_ID}' is ready.")
```

## 4. Upload Folder to Repository

`upload_folder()` recursively uploads everything inside `LOCAL_FOLDER` to the repo root (or a sub-path via `path_in_repo`).

```python
import os
from huggingface_hub import upload_folder

# Path to the local submission folder (FIXED for your system)
LOCAL_FOLDER = os.path.join(
    os.path.expanduser("~"),
    "Downloads",
    "Inference_MP1",
    "VRMP1_IMT2023035_IMT2023537_IMT2023550_IMT2023585"
)

# Commit message shown in the repo history
COMMIT_MESSAGE = "Add VRMP1 submission: predictor, validator, and model weights"

url = upload_folder(
    folder_path=LOCAL_FOLDER,
    repo_id=REPO_ID,
    repo_type=REPO_TYPE,
    commit_message=COMMIT_MESSAGE,
    ignore_patterns=["*.DS_Store", "__pycache__/*", "*.pyc"],
)

print("Upload complete!")
print("View your repo at:", url)
```

## 5. Verify Upload

List all files that were pushed to the repository to confirm everything landed correctly.

```python
files = api.list_repo_files(repo_id=REPO_ID, repo_type=REPO_TYPE)

expected = {
    "predictor.py",
    "validator_local.py",
    "Student_Submission_Guidelines.md",
    "model_files/cls.pth", # change this according to your actual format
    "model_files/seg.pth", # change this according to your actual format
}

uploaded = set(files)
print("Files in repository:")
for f in sorted(uploaded):
    status = "✓" if f in expected else " "
    print(f"  [{status}] {f}")

missing = expected - uploaded
if missing:
    print("\nWARNING — missing files:", missing)
else:
    print("\nAll expected files uploaded successfully!")
```
